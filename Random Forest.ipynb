```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y2015 = pd.read_csv('LoanStats3d.csv', error_bad_lines=False, skipinitialspace=True, header=1, use cols=["id"]
```

```python
y2015
```

```python
y2015.dropna(thresh=2)
y2015.dropna(axis=0, how='all')
y2015.dropna(axis=1, how='all')
y2015.columns = y2015.columns.str.replace(' ', '')
y2015
```

```python
y2015.iloc[:, ['term','grade', 'emp_length' ,'home_ownership' ,'verification_status' ,'issue_d ,loan_status' ,'pymnt_plan', 'purpose', 'title'
,'initial_list_status' , 'last_pymnt_d' , 'next_pymnt_d' , 'last_credit_pull_d' ,'application_type' , 'verification_status_joint' ]]
```

```python
# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
y2015.iloc[:, ['term','grade', 'emp_length' ,'home_ownership' ,'verification_status' ,'issue_d ,loan_status' ,'pymnt_plan', 'purpose', 'title'
,'initial_list_status' , 'last_pymnt_d' , 'next_pymnt_d' , 'last_credit_pull_d' ,'application_type' , 'verification_status_joint' ]]
```

```python
# Remove two summary rows at the end that don't actually contain data.
y2015 = y2015[:-2]
```

```python
categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
```
